```python
import pandas as pd
import lightgbm as lgb
import numpy as np

# -----------------------------
# Load data
# -----------------------------
train = pd.read_csv("/content/train_12m_features.csv")
future = pd.read_csv("/content/future_12m_features.csv")

# Convert 'object' columns with commas to numeric
def clean_numeric_columns(df, columns):
    for col in columns:
        if col in df.columns and df[col].dtype == 'object':
            df[col] = df[col].astype(str).str.replace(',', '', regex=False)
            df[col] = pd.to_numeric(df[col], errors='coerce')
    return df

numeric_cols = ["target", "lag_1m", "lag_12m", "rolling_mean_3m", "rolling_mean_12m"]
train = clean_numeric_columns(train, numeric_cols)
future = clean_numeric_columns(future, numeric_cols)

# Explicitly cast relevant columns to float to avoid FutureWarnings during assignments
train['target'] = train['target'].astype(float)
future['target'] = future['target'].astype(float)
future['lag_1m'] = future['lag_1m'].astype(float)
future['rolling_mean_3m'] = future['rolling_mean_3m'].astype(float)
future['rolling_mean_12m'] = future['rolling_mean_12m'].astype(float)

train["timestamp"] = pd.to_datetime(train["timestamp"])
future["timestamp"] = pd.to_datetime(future["timestamp"])

# -----------------------------
# Model features
# -----------------------------
FEATURES = ["month", "quarter", "season", "lag_1m", "lag_12m",
            "rolling_mean_3m", "rolling_mean_12m"]

# -----------------------------
# Train LightGBM
# -----------------------------
X_train = train[FEATURES]
y_train = train["target"]

model = lgb.LGBMRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=-1,
    num_leaves=20
)

model.fit(X_train, y_train)
print("Model trained.")

# -----------------------------
# Recursive monthly forecasting
# -----------------------------
future = future.sort_values("timestamp")

future_preds = []

# Combine to allow updating lag + rolling means
full_df = pd.concat([train, future], ignore_index=True)

for idx in future.index:

    row = future.loc[idx]

    # Extract values SAFELY (fixes your error)
    month   = int(row["month"])
    quarter = int(row["quarter"])
    season  = int(row["season"])
    lag_1m  = float(row["lag_1m"])
    lag_12m = float(row["lag_12m"])
    r3      = float(row["rolling_mean_3m"])
    r12     = float(row["rolling_mean_12m"])

    X_pred = pd.DataFrame([{
        "month": month,
        "quarter": quarter,
        "season": season,
        "lag_1m": lag_1m,
        "lag_12m": lag_12m,
        "rolling_mean_3m": r3,
        "rolling_mean_12m": r12
    }])

    # Predict
    yhat = float(model.predict(X_pred)[0])
    future_preds.append(yhat)

    # Write prediction back to full_df
    # Use full_df index, not future index
    full_df_idx = full_df[(full_df['group_id'] == row['group_id']) & (full_df['timestamp'] == row['timestamp'])].index[0]
    full_df.loc[full_df_idx, "target"] = yhat

    # Update lag_1m and rolling means for NEXT timestamp
    next_idx_in_future = future.index.get_loc(idx) + 1

    if next_idx_in_future < len(future.index):
        next_row = future.iloc[next_idx_in_future]
        # Get the actual index of the next row in the 'future' DataFrame
        next_future_original_idx = next_row.name

        future.loc[next_future_original_idx, "lag_1m"] = yhat

        grp = row["group_id"]
        # Ensure 'hist' only contains data up to the current prediction time for calculation
        hist = full_df[full_df["group_id"] == grp].sort_values("timestamp")
        hist_up_to_current = hist[hist['timestamp'] <= row['timestamp']]

        last3 = hist_up_to_current["target"].tail(3).mean()
        last12 = hist_up_to_current["target"].tail(12).mean()

        future.loc[next_future_original_idx, "rolling_mean_3m"] = last3
        future.loc[next_future_original_idx, "rolling_mean_12m"] = last12

# -----------------------------
# Save results
# -----------------------------
future["prediction"] = future_preds
future.to_csv("/content/monthly_forecasts.csv", index=False)

print("DONE! Saved /content/monthly_forecasts.csv")
```

```python
import pandas as pd
import lightgbm as lgb
import numpy as np

# -----------------------------
# Load data
# -----------------------------
train_h = pd.read_csv("/content/train_48h_features (1).csv")
future_h = pd.read_csv("/content/future_48h_features (1).csv")

# Convert 'object' columns with commas to numeric (re-defining for self-contained cell)
def clean_numeric_columns(df, columns):
    for col in columns:
        if col in df.columns and df[col].dtype == 'object':
            df[col] = df[col].astype(str).str.replace(',', '', regex=False)
            df[col] = pd.to_numeric(df[col], errors='coerce')
    return df

# Apply cleaning to train_h and future_h for 'target' column
train_h = clean_numeric_columns(train_h, ["target"])
future_h = clean_numeric_columns(future_h, ["target"])

train_h["timestamp"] = pd.to_datetime(train_h["timestamp"])
future_h["timestamp"] = pd.to_datetime(future_h["timestamp"])

# Explicitly cast relevant columns to float to avoid FutureWarnings during assignments
train_h['target'] = train_h['target'].astype(float)
future_h['target'] = future_h['target'].astype(float)

# -----------------------------
# Feature Engineering for hourly data
# -----------------------------
# Sort by group_id and timestamp before calculating lags and rolling means
train_h = train_h.sort_values(by=["group_id", "timestamp"])
future_h = future_h.sort_values(by=["group_id", "timestamp"])

# Extract time-based features
train_h["hour"] = train_h["timestamp"].dt.hour
train_h["dayofweek"] = train_h["timestamp"].dt.dayofweek

future_h["hour"] = future_h["timestamp"].dt.hour
future_h["dayofweek"] = future_h["timestamp"].dt.dayofweek

# Calculate lag features for training data
train_h["lag_1h"] = train_h.groupby("group_id")["target"].shift(1)
train_h["lag_24h"] = train_h.groupby("group_id")["target"].shift(24)

# Calculate rolling mean features for training data
train_h["rolling_mean_3h"] = train_h.groupby("group_id")["target"].transform(lambda x: x.rolling(3, closed='left').mean())
train_h["rolling_mean_24h"] = train_h.groupby("group_id")["target"].transform(lambda x: x.rolling(24, closed='left').mean())

# Initialize lag and rolling mean features for future data (will be updated recursively)
# These initial shifts will likely result in NaNs which will be filled in the recursive loop
future_h["lag_1h"] = future_h.groupby("group_id")["target"].shift(1)
future_h["lag_24h"] = future_h.groupby("group_id")["target"].shift(24)
future_h["rolling_mean_3h"] = future_h.groupby("group_id")["target"].transform(lambda x: x.rolling(3, closed='left').mean())
future_h["rolling_mean_24h"] = future_h.groupby("group_id")["target"].transform(lambda x: x.rolling(24, closed='left').mean())

# Ensure these columns are float type in future_h for recursive updates
future_h["lag_1h"] = future_h["lag_1h"].astype(float)
future_h["lag_24h"] = future_h["lag_24h"].astype(float)
future_h["rolling_mean_3h"] = future_h["rolling_mean_3h"].astype(float)
future_h["rolling_mean_24h"] = future_h["rolling_mean_24h"].astype(float)

# -----------------------------
# Features for hourly model (defined after feature engineering)
# -----------------------------
FEATURES_H = [
    "hour", "dayofweek",
    "lag_1h", "lag_24h",
    "rolling_mean_3h", "rolling_mean_24h"
]

# -----------------------------
# Train LightGBM model
# -----------------------------
X_train = train_h[FEATURES_H]
y_train = train_h["target"]

model_h = lgb.LGBMRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=-1,
    num_leaves=20
)

model_h.fit(X_train, y_train)
print("Hourly model trained.")

# -----------------------------
# Recursive 48h forecasting
# -----------------------------
# future_h is already sorted by timestamp from feature engineering steps

preds_48h = []

# Combine for updating lags & rolling means
full_hourly = pd.concat([train_h, future_h], ignore_index=True)

for idx in future_h.index:

    row = future_h.loc[idx]

    # Extract safely, handling potential NaNs by converting to 0.0 for prediction input
    # LightGBM can often handle np.nan, but converting to 0.0 is a common robust approach.
    hour   = int(row["hour"])
    dow    = int(row["dayofweek"])
    lag_1h = float(row["lag_1h"]) if pd.notna(row["lag_1h"]) else 0.0
    lag_24h = float(row["lag_24h"]) if pd.notna(row["lag_24h"]) else 0.0
    r3 = float(row["rolling_mean_3h"]) if pd.notna(row["rolling_mean_3h"]) else 0.0
    r24 = float(row["rolling_mean_24h"]) if pd.notna(row["rolling_mean_24h"]) else 0.0

    # Build prediction sample
    X_pred_h = pd.DataFrame([{
        "hour": hour,
        "dayofweek": dow,
        "lag_1h": lag_1h,
        "lag_24h": lag_24h,
        "rolling_mean_3h": r3,
        "rolling_mean_24h": r24
    }])

    # Predict
    yhat_h = float(model_h.predict(X_pred_h)[0])
    preds_48h.append(yhat_h)

    # Write prediction back to full_hourly
    # Find the corresponding index in full_hourly to update its 'target' with the prediction
    full_hourly_idx = full_hourly[(full_hourly['group_id'] == row['group_id']) & (full_hourly['timestamp'] == row['timestamp'])].index[0]
    full_hourly.loc[full_hourly_idx, "target"] = yhat_h

    # Update lag_1h, lag_24h and rolling means for NEXT timestamp in future_h
    next_idx_in_future = future_h.index.get_loc(idx) + 1

    if next_idx_in_future < len(future_h.index):
        next_row_in_future = future_h.iloc[next_idx_in_future]
        next_future_original_idx = next_row_in_future.name

        # Update features only if the next row belongs to the same group
        if next_row_in_future["group_id"] == row["group_id"]:
            future_h.loc[next_future_original_idx, "lag_1h"] = yhat_h

            grp = row["group_id"]
            # Ensure 'hist_hourly' only contains data up to the current prediction time for calculation
            hist_hourly = full_hourly[full_hourly["group_id"] == grp].sort_values("timestamp")
            hist_up_to_current_hourly = hist_hourly[hist_hourly['timestamp'] <= row['timestamp']]

            # Update rolling means for the next row
            if len(hist_up_to_current_hourly) >= 3:
                future_h.loc[next_future_original_idx, "rolling_mean_3h"] = hist_up_to_current_hourly["target"].tail(3).mean()
            else:
                future_h.loc[next_future_original_idx, "rolling_mean_3h"] = np.nan # Not enough data for rolling mean

            if len(hist_up_to_current_hourly) >= 24:
                future_h.loc[next_future_original_idx, "rolling_mean_24h"] = hist_up_to_current_hourly["target"].tail(24).mean()
            else:
                future_h.loc[next_future_original_idx, "rolling_mean_24h"] = np.nan # Not enough data for rolling mean

            # Update lag_24h for the next row
            lag_24h_timestamp = next_row_in_future['timestamp'] - pd.Timedelta(hours=24)
            lag_24h_value_series = hist_hourly[
                (hist_hourly['group_id'] == grp) &
                (hist_hourly['timestamp'] == lag_24h_timestamp)
            ]['target']

            if not lag_24h_value_series.empty:
                future_h.loc[next_future_original_idx, "lag_24h"] = lag_24h_value_series.iloc[0]
            else:
                future_h.loc[next_future_original_idx, "lag_24h"] = np.nan # 24h lag not found
        else: # If group_id changes, reset lags and rolling means for the new group
            future_h.loc[next_future_original_idx, "lag_1h"] = np.nan
            future_h.loc[next_future_original_idx, "lag_24h"] = np.nan
            future_h.loc[next_future_original_idx, "rolling_mean_3h"] = np.nan
            future_h.loc[next_future_original_idx, "rolling_mean_24h"] = np.nan

# -----------------------------
# Save results
# -----------------------------
future_h["prediction"] = preds_48h
future_h.to_csv("/content/hourly_forecasts.csv", index=False)

print("DONE! Saved /content/hourly_forecasts.csv")
```
